## Problem 2

Which of the following sets of functions are linearly dependent?

1. $\{\sin{x}\text{, }\cos{x}\text{, }x\sin{x}\}$

2. $\{\exp{x}\text{, }x\exp{x}\text{, }x^{2}\exp{x}\}$

3. $\{\sin^{2}{x}\text{, }\cos^{2}{x}\text{, }\cos{2x}\}$

### Problem 2 calculations

Suppose $(c_{1}, c_{2}, t) \in \mathbb{R}$.  From [LibreTexts Mathematics][1]:
<br>

>Let $f(t)$f and $g(t)$ be differentiable functions.  Then they are called *linearly dependent* if there are nonzero constants $c_1$ and $c_2$ with $c_{1}f(t) + c_{2}g(t) = 0$ for all $t$.  Otherwise they are called *linearly independent*.
<br>

For $n$ real or complex-valued functions $f_{1}, \ldots, f_{n}$, which are $n-1$ times differential over the interval $I$, the Wronskian can be used to systematically check for linear dependence.  The Wronskian is a fuction on $x\in I$ defined as:

$$
W(f_1,\ldots, f_n)(x) = 
\det
\begin{bmatrix}
  f_{1}(x) & f_{2}(x) & \ldots & f_{n}(x) \\
  f'_{1}(x) & f'_{2}(x) & \ldots & f'_{n}(x) \\
  \vdots & \vdots & \ddots & \vdots \\
  f_{1}^{(n-1)}(x) & f_{2}^{(n-1)}(x) & \ldots & f_{n}^{(n-1)}(x)
\end{bmatrix}
$$

If the Wronskian is nonzero for some $x$ in the intervale $I$, then the functions are linearly independent on $I$.  If the functions are linearly independent then the Wronskian is zero for all $x$ in $I$.  

:::{important}
If the Wronskian is 0 this does not necessarily imply that the functions are linearly dependent.
:::

1\. &emsp; $\{\sin{x}\text{, }\cos{x}\text{, }x\sin{x}\}$

Utilizing SymPy to define a python function that calculates the Wronskian for a set of functions:

[1]: <https://math.libretexts.org/Bookshelves/Analysis/Supplemental_Modules_(Analysis)/Ordinary_Differential_Equations/3%3A_Second_Order_Linear_Differential_Equations/3.6%3A_Linear_Independence_and_the_Wronskian> "Linear Independence and the Wronskian"

```python
# import necessary modules
from sympy import *

# define symbols
x = Symbol('x')

# define functions
f1 = sin(x)
f2 = cos(x)
f3 = x*sin(x)

# calculate first derivatives
d_f1 = f1.diff(x)
d_f2 = f2.diff(x)
d_f3 = f3.diff(x)

# calculate second derivatives
d2_f1 = d_f1.diff(x)
d2_f2 = d_f2.diff(x)
d2_f3 = d_f3.diff(x)

# create a symbolic matrix
sym_mat = Matrix([[f1, f2, f3], [d_f1, d_f2, d_f3], [d2_f1, d2_f2, d2_f3]])

# calculate the Wronskian
wronskian = sym_mat.det()

# print the result
print(f"The Wronskian is: {wronskian}")
```

The Wronskian for the system of equations is:

$$
W = -2\sin{x}^{2}\cos{x}-2\cos{x}^{3}
$$

Given this result is cyclic on the interval $I = [0, 2\pi{}]$, plotting the Wronskian over that domain will indicate if the result is nonzero.

Utilizing NumPy and Matplotlib (along with jtplot from jupyterthemes for styling) the plot of the Wronskian over the interval $I$ is shown below:

```python
# import necessary modules
import numpy as np
import matplotlib.pyplot as plt

# define domain and range
x = np.arange(0,2*np.pi, 0.001)
w = -2*np.sin(x)**2*np.cos(x)-2*np.cos(x)**3

# map the domain to the range
plt.plot(x, w)
plt.title("w(x) vs. x")
plt.xlabel("x")
plt.ylabel("w(x)")
plt.show()
```

The Wronskian is in fact zero over the interval.  Therefore more investigation is required to determine if the set of functions are linearly dependent or independent. Solving for when $w(x) = 0$:

Utilizing SymPy to solve for the roots of the polynomial:

```python
# import necessary modules
from sympy import *

# define the symbols
x = Symbol('x')

# define the equation
wronskian_eq = Eq(-2*sin(x)**2*cos(x) - 2*cos(x)**3, 0)

# solve the equation over the domain
sols = solve(wronskian_eq, x, domain=(0, 2*pi))

# print the results
print(f"Solutions: {sols}")
```

The roots are $\big\{ \frac{-\pi}{2}, \frac{\pi}{2} \big\}$, which is evident from the previous plot.

From inspection, it is clear that $\sin{x}$ and $\cos{x}$ are independent.  In fact they are orthogonal which is verifiable by computing the inner product $\langle \sin{x}, \cos{x} \rangle$ which will result in $0$.

This lends justification that $x\sin{x}$ and $\cos{x}$ are independent.  Turning our attention to the remaining pair, i.e., $\sin{x}$ and $x\sin{x}$, it is clear that when $x=0$ the pair is equivalent; however, there is no other combination of scalers (i.e., $a$ or $b$) other than $0$ that would make the pair equal to each other when $x \neq 0$.

:::{admonition} Answer
:class: tip
Therefore the set of functions are linearly independent.
:::

3\. &emsp; $\{\sin^{2}{x}\text{, }\cos^{2}{x}\text{, }\cos{2x}\}$

Utilizing SymPy to find the Wronkskian:

```python
# import necessary modules
from sympy import *

# define symbols
x = Symbol('x')

# define functions
f1 = sin(x)**2
f2 = cos(x)**2
f3 = sin(2*x)

# calculate first derivatives
d_f1 = f1.diff(x)
d_f2 = f2.diff(x)
d_f3 = f3.diff(x)

# calculate second derivatives
d2_f1 = d_f1.diff(x)
d2_f2 = d_f2.diff(x)
d2_f3 = d_f3.diff(x)

# create a symbolic matrix
sym_mat = Matrix([[f1, f2, f3], [d_f1, d_f2, d_f3], [d2_f1, d2_f2, d2_f3]])

# calculate the Wronskian
wronskian = sym_mat.det()

# print the result
print(f"The Wronskian is: {wronskian}")
```

The Wronskian for the system of equations is
<br>
<br>
$$
W = -4\sin^{4}{x}\cos{2x} + 8\sin^{3}{x}\sin{2x}\cos{x} + 8\sin{x}\sin{2x}\cos^{3}{x} + 4\cos^{4}{x}\cos{2x}
$$
<br>
Given this result is cyclic on the interval $I = [0, 2\pi{}]$, plotting the Wronskian over that domain will indicate if the result is nonzero.

Utilizing NumPy and Matplotlib (along with jtplot from jupyterthemes for styling) the plot of the Wronskian over the interval $I$ is shown below:

```python
# import necessary modules
import numpy as np
import matplotlib.pyplot as plt

# define domain and range
x = np.arange(0,2*np.pi, 0.001)
w = -4*np.sin(x)**4*np.cos(2*x) + 8*np.sin(x)**3*np.sin(2*x)*np.cos(x) + 8*np.sin(x)*np.sin(2*x)*np.cos(x)**3 + 4*np.cos(x)**4*np.cos(2*x)

# map the domain to the range
plt.plot(x, w)
plt.title("w(x) vs. x")
plt.xlabel("x")
plt.ylabel("w(x)")
plt.show()
```

The Wronskian is never zero over the interval.  In fact the Wronskian reduces to:

$$
W = 4
$$

The reduction is left to the reader as an exercise in trigonometric identities.

:::{admonition} Answer
:class: tip
Therefore the set of functions are linearly independent.

### Problem 3 calculations

From [this reference][2]:

> **Definition**: Let $V$ be a vector space over $F$, where $F = \mathbb{R}\text{ or }\mathbb{C}$. An inner product on $V$ is a function $\langle\cdot\text{,}\cdot\rangle\text{: }V \times{} V\rightarrow F$ which satisifies:
<br>
>
> 1. Linearity: $\langle{}ax+y\text{, }z\rangle = a\langle{}x\text{, }z\rangle + \langle{}y\text{, }z\rangle\text{ for }x\text{, }y\text{, z}\in V\text{, }a \in F$
>
> 2. Symmetry: $\langle{}x\text{, }y\rangle = \overline{\langle{}y\text{, }x\rangle}\text{, for }x\text{, }y \in V$
<br>
> 3. Positive-definiteness: If $x\neq 0\text{, then }\langle{}x\text{, }x\rangle \in \mathbb{R} \text{ and }\langle{}x\text{, }x\rangle\gt 0$
<br> 

1\. &emsp; $\langle\mathbf{x}\text{, }\mathbf{y}\rangle = x_{1}y_{1} + x_{3}y_{3}$

Check for linearity in the first slot, we don't have to check the second slot because we'll check for symmetry if this property holds:

Since $a\langle{}x\text{, }y\rangle = \langle{}ax\text{, }y\rangle$ to ensure linearity of the entire inner product space...

$$
\begin{align}
  \langle{}ax\text{, }y\rangle & = a(x_{1}y_{1} + x_{3}y_{3})\\
  & = ax_{1}y_{1} + ax_{3}y_{3}
\end{align}
$$

As long as the property $a\langle{}x\text{, }y\rangle = \langle{}ax\text{, }y\rangle$ holds true, then scaler multiplication is satisifed.

Checking for additivity:

$$
\begin{align}
  \langle{}x_1 + x_2\text{, }y\rangle &= (x_1 + x_2)y_1 + x_{3}y_{3}\\
  & = x_{1}y_{1} + x_{2}y_{1} + x_{3}y_{3}\\
\end{align}
$$

From the inner product property $\langle{}x_1 + x_2\text{, }y\rangle = \langle{}x_1\text{, }y\rangle{} + \langle{x_2}\text{, }y\rangle$

$$
\langle{}x_1\text{, }y\rangle + \langle{}x_2\text{, }y\rangle = x_{1}y_{1} + x_{3}y_{3} + x_{2}y_{1} + x_{3}y_{3}
$$

Since:

$$
\begin{align}
  x_{1}y_{1} + x_{2}y_{1} + x_{3}y_{3} & \neq x_{1}y_{1} + x_{3}y_{3} + x_{2}y_{1} + x_{3}y_{3}\\
  \langle{}x_1 + x_{2}\text{, }y\rangle & \neq \langle{}x_1\text{, }y\rangle + \langle{}x_{2}\text{, }y\rangle
\end{align}
$$

:::{admonition} Answer
:class: tip
Therefore this is not an inner product space for $\mathbb{R}^3$.
:::

Checking for additivity:

$$

  & = y_{1}x_1 - y_{2}x_2 + y_{3}x_3
\end{align}
$$

Therefore symmetry is satisifed.

Checking for positive definiteness:

$$
\langle{}x\text{, }x\rangle = x_{1}^2 - x_{2}^2 + x{3}^2
$$

If $x_{2}^2 \gt x_{1}^{2} + x_{3}^2$ then the inner product is not positive definite.

:::{admonition} Answer
:class: tip
Therefore this is not an inner product space for $\mathbb{R}^3$.
:::

3\. &emsp; $\langle\mathbf{x}\text{, }\mathbf{y}\rangle = 2x_{1}y_{1} + x_{2}y_{2} + 4x_{3}y_{3}$

Checking for scaler multiplicity:

$$
\begin{align}
  \langle{}ax\text{, }y\rangle & = 2ax_{1}y_1 + ax_{2}y_2 + 4ax_{3}y_3\\
  & =a\langle{}x\text{, }y\rangle \\
  & = a(2x_{1}y_1 + x_{2}y_2 + 4x_{3}y_3)
\end{align}
$$

Therefore scaler multiplicity is satisfied.

Checking for additivity:

$$
\begin{align}
  \langle{}x_1 + x_2\text{, }y\rangle & = 2(x_1 + x_2)y_1 + (x_1 + x_2)y_2 + 4(x_1 + x_2)y_3\\

  & = 2x_{1}y_1 + 2x_{2}y_1 + x_{1}y_2 + x_{2}y_2 + 4x_{1}y_3 + 4x_{2}y_3
\end{align}
$$

Therefore additivity is satisifed.

Checking for symmetry:

$$
\begin{align}
  \langle{}x\text{, }y\rangle & = 2x_{1}y_1 + x_{2}y_2 + 4x_{3}y_3\\
    & = \langle{}y\text{, }x\rangle \\
  & = 2y_{1}x_1 + y_{2}x_2 + 4y_{3}x_3
\end{align}
$$

Therefore symmetry is satisifed.

Checking for positive definiteness:

\langle{}x\text{, }x\rangle = 2x_{1}^2 + x_{2}^2 + 4x_{3}^2
$$

This expression is equal to 0 only when $x_1 = x_2 = x_3 = 0$ and otherwise is always nonnegative so positive definiteness is satisifed.

:::{admonition} Answer

Therefore this is an inner product space for $\mathbb{R}^3$.

## Problem 4

Show the following basis functions are orthogonal to each other in the interval $x \in [-\pi\text{, }\pi]$.  In the following $m$ and $n$ are nonzero integer numbers.

2. Fourier modes of $\{\cos{(mx)}\}$ and $\{\cos{(nx)}\}$ when $n \neq m$.

3. Fourier modes of $\{\sin{(mx)}\}$ and $\{\cos{(nx)}\}$.

### Problem 4 calculations

For these problems, we need to take the inner product of the functions over the given interval and show that under the given conditions this result is equal to zero thus demonstrating orthogonality.

For all three sub-problems I'll be using this python code from the SymPy module to make my life a little easier.

```python
# import necessary modules
from sympy import *

def ip_func(f1, f2):
    # define symbols
    x, m, n = symbols('x m n')

    # calculate the inner product symbolically
    ip_result = integrate(f1 * f2, (x, -pi, pi))
    ip_simplified = simplify(ip_result)
    
    # return the result
    return ip_simplified
```

1\. &emsp; Fourier modes of {$\sin{(mx)}$} and {$\sin{(nx)}$} when $n \neq m$.

Evaluate the inner product over the interval:

$$
\langle{}\sin{(mx)}\text{, }\sin{(nx)}\rangle = \int\limits_{-\pi}^\pi\sin{(mx)}\sin{(nx)}\,dx
$$

```python
# import necessary modules
from sympy import *

x, m, n = symbols('x m n')

# define the functions
f1 = sin(m * x)
f2 = sin(n * x)

# evaluate the inner product
ip_func(f1, f2)
```

The last term in the piece-wise expression is what we'll evaluate because that complies with the problem constraints (i.e., $m\text{, } n \in \mathbb{Z} \text{, } m\text{, }n \gt 0 \text{, and }m\neq n$).

```python
# define abritrary values for m and n subject to constraints
n = 1
m = 2

# plug and chug
print(2*(-m*sin(pi*n)*cos(pi*m) + n*sin(pi*m)*cos(pi*n))/(m**2 - n**2))
```

:::{admonition} Answer
:class: tip
The inner product is zero therefore the Fourier modes are orthogonal subject to the constraints.
:::
